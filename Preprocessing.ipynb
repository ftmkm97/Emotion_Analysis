```python
import pandas as pd
import numpy as np
import re

pd.options.mode.chained_assignment = None  
```

# Feature Selection

```python
df = pd.read_csv('emotional.csv')
df
```

```python
df.info()
```

```python
df = df[['Tags','Author Username','Times Viewed','Rating','Created At', 'Category']]
df
```

# Preprocessing

```python
df['Tags'] = df['Tags'].str.replace(',',' ')
```

```python
df['Tags'] = df['Tags'].str.lower()
```

### Removing Emojies

```python
def deEmojify(inputString):
    return inputString.encode('ascii','ignore').decode('ascii')

df['Tags']= df['Tags'].apply(lambda s: deEmojify(s))
df
```

### Removing HTML Tags

```python
def remove_html(text):
    html_pattern = re.compile('<.*?>')
    return html_pattern.sub(r' ', text)


   
df['Tags']= df['Tags'].apply(lambda h: remove_html(h))
df
```

### Removing URLs

```python
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r' ', text)

df['Tags']= df['Tags'].apply(lambda r: remove_urls(r))
df
```

### Removing Email Addresses

```python

def remove_Email(text):
    url_pattern = re.compile(r'[\w\.-]+@[\w\.-]+')
    return url_pattern.sub(r' ', text)


df['Tags']= df['Tags'].apply(lambda r: remove_Email(r))
df
```

### Removing Social Media IDs

```python
def remove_id(text):
    url_pattern = re.compile(r'@\w')
    return url_pattern.sub(r' ', text)

df['Tags']= df['Tags'].apply(lambda r: remove_id(r))
df
```

### Removing Numbers 

```python
df["Tags"] = df["Tags"].replace('\d+', '', regex=True)
df
```

### Removing Punctuations

```python
import re

df["Tags"] = df["Tags"].apply(lambda s: ' '.join(re.sub("[.,!?:;-='...\"@#_]", " ", s).split()))
df
```

### Removing Stopwords

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords


stop = stopwords.words('english')


df['Tags'] = df['Tags'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
df
```

### Removing Non-English Words

```python
import nltk
nltk.download('words')
words = set(nltk.corpus.words.words())

df['Tags'] = df['Tags'].apply(lambda x: " ".join(w for w in nltk.wordpunct_tokenize(x) if w.lower() in words or not w.isalpha()))

df
```

### Remove Repetitive Characters

```python
def conti_rep_char(str1):
    tchr = str1.group(0)
    if len(tchr) > 1:
        return tchr[0:1]
     

    
def check_unique_char(rep, sent_text):
    convert = re.sub(r'(\w)\1+', 
                     rep,
                     sent_text)
    return convert
 
df['Tags'] = df['Tags'].apply(lambda x : check_unique_char(conti_rep_char, x))

df
```

### Stemming


```python
from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer("english")


def stim_text(text):
    st = [stemmer.stem(w) for w in text.split()]
    st = " ".join(st)
    return(st)


df['Tags'] = df['Tags'].apply(lambda s: stim_text(s))

df
```
